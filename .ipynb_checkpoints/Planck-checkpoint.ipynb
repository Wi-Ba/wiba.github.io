```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider, Checkbox, VBox
from IPython.display import Markdown, display
```

```python
class BlackBody:

    h = 6.62607015e-34
    c = 2.99792458e8
    k = 1.380649e-23
    sigma = 5.670374419e-8

    def __init__(self, temperature):
        self.T = temperature

    def planck(self, wavelength):
        """
        Planck law

        wavelength in meters
        """
        h = self.h
        c = self.c
        k = self.k
        T = self.T

        a = 2 * np.pi * h * c**2
        b = h * c / (wavelength * k * T)

        return a / (wavelength**5 * (np.exp(b) - 1))

    def wien_peak(self):
        """
        Peak wavelength in meters
        """
        b = 2.897771955e-3
        return b / self.T

    def total_emissive_power(self):
        """
        Stefan-Boltzmann law
        """
        return self.sigma * self.T**4
```

```python
theory = r"""
# Blackbody Radiation

## Planck's Law

Planck's law describes the spectral emissive power of a blackbody:

\[
E_\lambda=
\frac{2\pi hc^2}
{\lambda^5
\left(
e^{hc/(\lambda kT)}-1
\right)}
\]

where

- \(h\) = Planck constant
- \(c\) = speed of light
- \(k\) = Boltzmann constant
- \(T\) = temperature

---

## Wien's Displacement Law

The wavelength at which the emissive power reaches its maximum is

\[
\lambda_{max} T
=
2.898\times10^{-3}
\; \text{m K}
\]

As temperature increases, the peak shifts to smaller wavelengths.

---

## Stefan-Boltzmann Law

The total emissive power is

\[
E = \sigma T^4
\]

where

\[
\sigma = 5.670\times10^{-8}
\;W/m^2K^4
\]

This explains why the area under the Planck curve increases rapidly with temperature.
"""

display(Markdown(theory))
```

```python
def plot_blackbody(T=1000,
                   show_reference=False,
                   T_ref=2000, 
                   log_scale=False):

    
    peak_main = 2898 / T      # μm
    peak_ref  = 2898 / T_ref  # μm

    peak_min = min(peak_main, peak_ref)
    peak_max = max(peak_main, peak_ref)

    xmin = max(0.1, peak_min / 8)
    xmax = peak_max * 8
    
    wavelength_um = np.linspace(xmin, xmax, 2000)
    wavelength_m = wavelength_um * 1e-6

    bb = BlackBody(T)

    E = bb.planck(wavelength_m)

    plt.figure(figsize=(10,6))

    plt.plot(
        wavelength_um,
        E,
        lw=3,
        label=f"{T:.0f} K"
    )

    
    plt.fill_between(
        wavelength_um,
        E,
        alpha=0.25,
        color='skyblue'
    )

    if show_reference:

        ref = BlackBody(T_ref)

        plt.plot(
            wavelength_um,
            ref.planck(wavelength_m),
            '--',
            color='gray',
            lw=2,
            label=f"Reference: {T_ref:.0f} K"
        )

    peak = bb.wien_peak()*1e6
    peak_index = np.argmax(E)

    plt.plot(
        wavelength_um[peak_index],
        E[peak_index],
        'ro',
        markersize=8
        )

    plt.axvline(
        peak,
        color='red',
        linestyle=':',
        label=f"Wien Peak = {peak:.2f} μm"
    )

    if log_scale:
        plt.yscale("log")
        
    plt.xlabel("Wavelength (μm)")
    plt.ylabel("Spectral Emissive Power")
    plt.title("Planck Distribution")

    
    # Visible spectrum
    plt.axvspan(
    0.4,
    0.7,
    alpha=0.15,
    color="green",
    label="Visible spectrum"
    )

    
    plt.grid(True)
    plt.legend()

    plt.show()

    print(f"Peak wavelength = {peak:.3f} μm")
    print(f"Total emissive power = {bb.total_emissive_power():.3e} W/m²")
```

```python
interact(
    plot_blackbody,

    log_scale=Checkbox(
        value=False,
        description="Log scale"
    ),

    T=FloatSlider(
        value=1000,
        min=300,
        max=6000,
        step=50,
        description="T (K)"
    ),

    show_reference=Checkbox(
        value=False,
        description="Reference"
    ),

    T_ref=FloatSlider(
        value=2000,
        min=300,
        max=6000,
        step=50,
        description="Ref T"
    )
    
);
```
